```python
# Drop rows where 'title' column is missing
drop_missing_titles = df.dropna(subset=['title'])
drop_missing_titles.isnull().sum()
```

```python
drop_missing_titles.shape
```

```python
# sum rows where 'title' column is duplicated
df.duplicated('title').sum()

```

```python
# drop rows where 'id' column is duplicated
df_missing_title_dup = df.drop_duplicates('id');
df_missing_title_dup.duplicated('id').sum()
```

```python
df_missing_title_dup.shape
```

```python
df_missing_title_dup.isnull().sum().sum()
```

```python
#replace all missing data with 0
replace_empty_values_with_null = df_missing_title_dup.fillna(value = 'null')
```
